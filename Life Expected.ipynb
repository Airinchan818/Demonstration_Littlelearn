```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("nailasrivastava/life-expectancy-analysis")

print("Path to dataset files:", path)
```

```python
import pandas as pd
datasets = pd.read_csv(path + "/Life Expectancy Data.csv")
```

```python
datasets
```

```python
x_data = datasets.copy()
```

```python
x_data = x_data.drop("Country", axis=1)
x_data = x_data.drop("Year",axis=1)
x_data = x_data.drop("Status",axis=1)
x_data = x_data.drop("infant deaths",axis=1)
```

```python
x_data
```

```python
from littlelearn.preprocessing import MinMaxScaller
```

```python
x_data.dropna(inplace=True)
```

```python
y_target = x_data['Life expectancy '].values
x_data = x_data.drop("Life expectancy ",axis=1)
```

```python
scalers_population = MinMaxScaller(f_range=(0,100))
scalers_GDP = MinMaxScaller(f_range=(0,1))
scalers_measles = MinMaxScaller(f_range=(0,10))

scalers_population.fit(x_data["Population"].values.reshape(-1,1))
scalers_GDP.fit(x_data["GDP"].values.reshape(-1,1))
scalers_measles.fit(x_data["Measles "].values.reshape(-1,1))
```

```python
x_data['Population']=scalers_population.scaling(x_data['Population'].values.reshape(-1,1))
x_data['GDP']=scalers_GDP.scaling(x_data['GDP'].values.reshape(-1,1))
x_data['Measles ']=scalers_measles.scaling(x_data['Measles '].values.reshape(-1,1))
```

```python
x_data
```

```python
x_train = x_data.values
```

```python
y_target = y_target.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(x_train,y_target,test_size=0.2,random_state=42)
```

```python
import littlelearn.DeepLearning as dl
```

```python
Model = dl.Model.Sequential([
    dl.layers.Dense(256,activation=dl.activations.Leaky_Relu()),
    dl.layers.Dense(128,activation=dl.activations.Leaky_Relu()),
    dl.layers.Dense(256,activation=dl.activations.Leaky_Relu()),
    dl.layers.Dense(64,activation=dl.activations.Leaky_Relu()),
    dl.layers.Dense(1,activation=dl.activations.Linear()),
])
```

```python
Model.detail()
```

```python
Model.build_model(optimizer=dl.optimizers.Adam(),
                  loss = dl.loss.MeanAbsoluteError())
```

```python
Model.fit(x=X_train,y=Y_train,epochs=500,verbose=1)
```

```python
Model.plot_loss()
```

```python
Model.plot_graph_execution()
```

```python
y_pred = Model.predict(X_test)
```

```python
dl.loss.MeanAbsoluteError()(Y_test,y_pred)
```
